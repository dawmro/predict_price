```python
import yfinance as yf
```

```python
# Download price history for GSPC
sp500 = yf.Ticker("^GSPC")
```

```python
# Query all historical data
sp500 = sp500.history(period="max")
```

```python
sp500
```

```python
# Show closing price for each day
sp500.plot.line(y="Close", use_index=True)
```

```python
# Remove useless columns
if "Dividends" in sp500:
  del sp500["Dividends"]
if "Stock Splits" in sp500:
  del sp500["Stock Splits"]
```

```python
sp500
```

```python
# Tomorrow price
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
# Check if tomorrow price is higher than today, it will be later use for predictions
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
sp500
```

```python
# Use only data after market shifted into current trend
sp500 = sp500.loc["1990-01-01":].copy()
```

```python
sp500
```

```python
## Training initial machine learning model
# Import favourite model for machine learning
from sklearn.ensemble import RandomForestClassifier

# Initialize model
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

# Split data into train and test set
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

# Make list with columns that will be used to predict the target
predictors = ["Close", "Volume", "Open", "High", "Low"]

# Use predictors to try to predict target
model.fit(train[predictors], train["Target"])
```

```python
# Measure how accurate model is
from sklearn.metrics import precision_score
preds = model.predict(test[predictors])
```

```python
# Turn Numpy array into Pandas series
import pandas as pd
preds = pd.Series(preds, index=test.index)
```

```python
# Calculate precision score
precision_score(test["Target"], preds)
```

```python
# Plot predictons by combining actual values with predicted values, use each input as colums
combined = pd.concat([test["Target"], preds], axis=1)
combined.plot(figsize=(20,5))
```

```python
## Backtesting system
# Prediction function
def predict(train, test, predictors, model):
  model.fit(train[predictors], train["Target"])
  preds = model.predict(test[predictors])
  preds = pd.Series(preds, index=test.index, name="Predictions")
  combined = pd.concat([test["Target"], preds], axis=1)
  return combined
```

```python
# Backtest function, every year has 250 traiding days,
# take first 10 years of data and predict values for 11th year,
# next take first 11 years of data and predict values for 12th year and so on...
def backtest(data, model, predictors, start=2500, step=250):
  # List of dataframes, one dataframe is predictions for single year
  all_predictions = []
  # Loop across data year by year
  for i in range(start, data.shape[0], step):
    # Split data into train and test set
    # Train set is years before current year
    train = data.iloc[0:i].copy()
    # Test set is current year
    test = data.iloc[i:(i+step)].copy()
    # Generate predictions
    predictions = predict(train, test, predictors, model)
    # Append predictions for current year
    all_predictions.append(predictions)
  # Return concatenated predictions into single dataframe
  return pd.concat(all_predictions)


```

```python
# Backtest data with model and predictors
predictions = backtest(sp500, model, predictors)
```

```python
# How many days has been predicted to go down(0) vs up(1)
predictions["Predictions"].value_counts()
```

```python
# Calculate precision score
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
# Compare to days when market went up
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
# Mean prices in following time horizons
horizons = [2,5,60,350,1000]
new_predictors = []

# Calculate rolling average in a given horizon
for horizon in horizons:
  rolling_averages = sp500.rolling(horizon).mean()

  # Add it to sp500 dataset
  ratio_column = f"Close ratio {horizon}"
  sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]

  # Check number of days in the laxt X days when stock went up
  trend_column = f"Trend {horizon}"
  sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

  new_predictors += [ratio_column, trend_column]

```

```python
# Remove rows with missing data
sp500 = sp500.dropna()
```

```python
## Improve model
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)

```

```python
# Improve Prediction function
def predict(train, test, predictors, model):
  model.fit(train[predictors], train["Target"])
  # Get probability that stock goes up
  preds = model.predict_proba(test[predictors])[:,1]
  # Set custom treshold to 60%
  preds[preds >= 0.6] = 1
  preds[preds < 0.6] = 0
  preds = pd.Series(preds, index=test.index, name="Predictions")
  combined = pd.concat([test["Target"], preds], axis=1)
  return combined
```

```python
# Run backtest
predictions = backtest(sp500, model, new_predictors)
```
